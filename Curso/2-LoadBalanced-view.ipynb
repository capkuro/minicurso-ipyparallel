```python
from ipyparallel import Client
rc = Client(profile='pucv-cluster')
len(rc)
```

```python
rc.purge_everything() #Limpiamos la sesion
```

```python
bview = rc.load_balanced_view(targets=list(range(10)))
```

```python
import numpy as np
```

```python
@bview.parallel()
def get_a(placeholder):
    print(placeholder)
    return placeholder
```

```python
res = get_a.map(np.random.rand(200,100))
```
